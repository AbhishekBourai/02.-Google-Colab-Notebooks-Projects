#Example

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# 3 times plot meaning 3 lines will be plotted in same graph 

plt.plot(ages, py_salaries, label='Python')
plt.plot(ages, js_salaries, label='JavaScript')

plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

# Figure : Container holding our plot and 
#Axis: is the actual plot 

**we use more Object Oriented approach when using Subplots**

```python
plt.style.use('seaborn')

ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Plotting same graph with Object Oriented approach  
# we can define certain no of rows and columns for the plots ; by default it is 1
# 'ax' is nothing bu the actual plot so instead of plt.plot() use--> ax.plot()
# plt.subplots() is a method returning the Figure and Axis(Plots)[the number is set by us]
fig, ax = plt.subplots()

ax.plot(ages, py_salaries, label='Python')
ax.plot(ages, js_salaries, label='JavaScript')

ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

# But when specifying the Title and X and Y labels- we use set_xlabel and set_ylabel etc. 
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

# Now defining multiple rows and columns

printing 'ax' Showing single plot --> since we haven't defined any multiple rows/columns

```python
plt.style.use('seaborn')

fig, ax = plt.subplots()

print(ax)

```

```python
# Multiple rows and columns

plt.style.use('seaborn')

fig, ax = plt.subplots(nrows=2, ncols=1)

print(ax)
```

```python
# Multiple rows and columns- 2 rows and 2 columns

plt.style.use('seaborn')

fig, ax = plt.subplots(nrows=2, ncols=2)

print(ax)
```

But the above plots will still be in the same figure hence in order to plot them separately we need to unpack like below.

```python
# Multiple rows and columns- 2 rows and 2 columns

plt.style.use('seaborn')

# Unpacking - Now we will have 2 separate plots
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

print(ax1)
print(ax2)
```

```python
plt.style.use('seaborn')

ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']


fig, ax1 = plt.subplots()

ax1.plot(ages, py_salaries, label='Python')
ax1.plot(ages, js_salaries, label='JavaScript')

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.style.use('seaborn')

ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']


fig, (ax1, ax2) = plt.subplots(nrows=2,ncols=1)

ax1.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_title('Median Salary (USD) by Age')
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
plt.style.use('seaborn')

ages = data['Age']

dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

# Share x axis by sharex
fig, (ax1, ax2) = plt.subplots(nrows=2,ncols=1,sharex=True)

ax1.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax2.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

# removed repetitive legends
ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.plot(ages, py_salaries, label='Python')
ax2.plot(ages, js_salaries, label='JavaScript')

ax1.legend()
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()
ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()

plt.show()

fig1.savefig('fig1.png')
fig2.savefig('fig2.png')
```
